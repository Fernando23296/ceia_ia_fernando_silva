##### 3.1 Empiezo con la Regresión Lineal simple

```python
def normalizador(columna):
    columna_norm = (columna - np.mean(columna))/np.std(columna)
    return columna_norm
```

```python
def separador(columna):

    m = columna.shape[0]
    train_size = (70*m)//100
    val_size = (20*m)//100
    
    train = columna[0:train_size]
    val = columna[train_size:train_size + val_size]
    test= columna[train_size + val_size:]
    
    return(train, val, test)
```

```python
class Regresion:
    def __init__(self, grados = 1): 
        self.grados = grados

    def entrenamiento(self, X, y):
        A = []
        B = []
        for i in range(self.grados):
            A_row = []
            for j in range(self.grados):
              potencia_x = []
              for x in X:
                potencia_x.append(x**(i+j))

              A_row.append(np.sum(potencia_x))

            A.append(A_row)
            extra_2 = []
            
            for x in X:
              extra_2.append(x**i)

            extra_3 = np.multiply(extra_2, y)

            B.append(np.sum(extra_3))
            
        w = np.linalg.inv(A) * np.transpose(B)
        self.w = w
    
    def prediccion(self, X): 
        pred = [] 
        for x in X:
            extra_3 = []
            for j, w_j in enumerate(self.w):
                extra_3.append(w_j * x**j)
            pred.append(np.sum(extra_3)) 
        return pred
    
```

```python
X = normalizador(X)
Y = normalizador(Y)

X_train, X_test, X_val = separador(X)
Y_train, Y_test, Y_val = separador(Y)
```

```python
x_min = min(X)
x_max = max(X)
x_nuevo = np.arange(x_min, x_max, 0.05)
y_nuevo = np.sin(x_nuevo) + np.random.rand(x_nuevo.shape[0])
print(x_nuevo.shape)
```

```python
model = Regresion(grados=2)
model.entrenamiento(X, Y)
plt.scatter(X, Y)
plt.plot(x_nuevo, model.prediccion(x_nuevo), c="red")
plt.show()
```

#### 4. Obtener mediante cross-validation para cada polinomio el error de validación.

```python
def mse(y_true, y_pred):
    A = np.sum((y_true - y_pred) ** 2)
    mse_ = A/y_true.shape[0]
    return mse_
```

```python
mse_data = []
m  = 15
for i in range(1,m):
    model = Regresion(grados=i)
    model.entrenamiento(X, Y)
    y_pred = model.prediccion(X)
    mse_data.append(mse(Y,y_pred))
    
```

```python
plt.plot(num_data, mse_data)
plt.show()
```
