### Función Sigmoid

Implementar la función: $g(z) = \frac{1}{1 + e^{-z}}$ en NumPy

```python
# TODO
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

### Binary Cross Entropy

Implementar la función de costo: $J(w) = \frac{1}{n}\sum_{i=1}^{n}L\left ( \hat{y},y \right )= \frac{1}{n}\sum_{i=1}^{n}\left [y^{(i)}log(\hat{y}^{(i)})+ (1-y^{(i)})log(1-\hat{y}^{(i)}) \right ]$

```python
# TODO
def bce(X, y, theta):
    h = sigmoid(np.dot(X,theta))
    cost = (1/len(y))*((np.dot((-y).T,np.log(h)))- (np.dot((1-y).T, np.log(1-h))))
    return cost
```

### Gradiente

Implementar el gradiente de la función costo respecto de los parámetros: $\frac{\partial J(w)}{\partial w} = \frac{1}{n}\sum_{i=1}^{n}\left ( \hat{y}^{i}-y^{i}\right )\bar{x}^i$

```python
# TODO
def gradient_descent(X, y, p, lr, epochs):
    hist_cos = np.zeros((epochs,1))
    for i in range(epochs):
        p = p - (lr/len(y)) * np.dot(X.T,(sigmoid(np.dot(X,p)) - y)) 
        hist_cos[i] = bce(X, y, p)

    return (hist_cos, p)
```

### Normalización

Implementar normalización Z-score de las features de entrada

```python
# TODO
def normalization(x):
    n_mean = np.mean(x)
    n_std = np.std(x)
    x_norm = np.array([(i - n_mean)/n_std for i in x])
    
    return x_norm
```

### Métricas (Precision, Recall y Accuracy)

Implementar las métricas en NumPy

```python
# TODO
class Metricas:
    def __init__(self, truth, prediction):
        self.truth = truth
        self.prediction = prediction
    
    def boolean_list(self):
            l = []
            for i in range (0,len(self.truth)):
                if (self.truth[i]==1) & (self.prediction[i]==1):
                    l.append("TP")
                elif (self.truth[i]==0) & (self.prediction[i]==0):
                    l.append("TN")
                elif (self.truth[i]==1) & (self.prediction[i]==0):
                    l.append("FN")
                else:
                    l.append("FP")
            return (l)

    def variables(self):
        bool_l = self.boolean_list()
        l_variables = ["TP", "TN", "FN", "FP"]
        l_variables_count = []
        for i in l_variables:
            l_variables_count.append(bool_l.count(i))
        return (l_variables_count)
        
    def precision(self):
        l = self.variables()
        res = l[0] / (l[0] + l[3])
        return (res)
    
    def recall(self):
        l = self.variables()
        res = l[0] / (l[0] + l[2])
        return (res)
    
    def accuracy(self):
        l = self.variables()
        res = (l[0] + l[1])/ (l[0] + l[1] + l[2] + l[3])
        return (res)
```
