### 1. Aplicando PCA

#### 1.1 PCA en Numpy

```python
X = X - np.mean(X, axis = 0) 
```

```python
cov = np.cov(X, rowvar = False)
```

```python
eigen_val , eigen_vec = np.linalg.eig(cov)
```

```python
idx = np.argsort(eigen_val)[::-1]
```

```python
eigen_vecs = eigen_vec[:,idx]
eigen_vals = eigen_val[idx]

a = np.dot(X, eigen_vec)
a = a * (-1)
```

```python
pca_2d = a[:,[0,1]]
```

```python
pca_2d.shape
```

```python
eigen_vals.shape
```

### 2. Graficar un scree plot (varianza contemplada en funci칩n del n칰mero de componentes considerados)

```python
num_eig_vals = eigen_vals[:10]
sum_eigenvalues = np.sum(num_eig_vals)

prop_var = [i/sum_eigenvalues for i in num_eig_vals]
cum_var = [np.sum(prop_var[:i+1]) for i in range(len(prop_var))]


import matplotlib.pyplot as plt

x_labels = ['PC{}'.format(i+1) for i in range(len(prop_var))]

plt.plot(x_labels, prop_var, marker='o', markersize=6, color='blue', linewidth=2, label='Proporci칩n de la varianza')
plt.plot(x_labels, cum_var, marker='o', color='red', linewidth=2, label="Varianza Acumulada")
plt.legend()
plt.title('Scree plot')
plt.xlabel('N췈 Components')
plt.ylabel('Varianza contemplada')
plt.show()
```
